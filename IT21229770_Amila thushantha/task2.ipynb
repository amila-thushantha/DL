```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv('Au_nanoParticle_dataset.csv')
```

```python
# Filter the required columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
# Display the first 20 samples
print(new_df.head(20))
```

```python
# Display the first 20 samples
print(new_df.head(20))
```

```python
# Calculate statistics
mean = new_df.mean()
std_dev = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

# Display the statistics
print("Mean:\n", mean)
print("\nStandard Deviation:\n", std_dev)
print("\nQuartiles:\n", quartiles)
```

```python
# Plot histograms
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for i, feature in enumerate(features):
    axes[i].hist(new_df[feature], bins=20, alpha=0.7)
    axes[i].set_title(feature)
plt.tight_layout()
plt.show()
```

```python
# Pairplot using seaborn
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)  
g.map_diag(sns.histplot, kde=True) 
g.map_lower(sns.kdeplot) 
plt.show()
```
