Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

create random 4x4 array

```python
random_array = np.random.exponential(scale=1.0, size=(4, 4))
print(random_array)
```

create exponential random array 100000x1

```python
exp_array = np.random.exponential(scale=1.0, size=(100000, 1))
uniform_array = np.random.rand(100000)
normal_array = np.random.randn(100000)
```

```python
exp_array
```

```python
plt.hist(exp_array, density=True, bins=100, histtype="step", color="red", label="Exponential")
```

plotting and visualizing the graph

```python
plt.hist(exp_array, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_array, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.axis([-2.5, 10, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

import matplotlib and plot in 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

visualizing the plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

import pokeman dataset

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

define columns

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]
```

calculate correlation

```python
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')
```

visualizing results using heatmaps

```python
plt.figure(figsize=(10, 5))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation')
plt.show()

plt.figure(figsize=(10, 5))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation')
plt.show()
```
